```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
gni = pd.read_csv('data/gni_per_cap_atlas_method_con2021.csv', header=[0], index_col=[0])
gni.index.name = 'Year'  
gni.fillna(0, inplace=True)
gni1 = gni.transpose()
gni2 = gni1.replace({'k': '*1e3', 'M': '*1e9' }, regex=True).map(pd.eval).astype(int)
years = list(gni2.index)
north_america = gni2[['United States','Mexico','Canada']]
before_nafta = north_america[173:194]
after_nafta = north_america[194:215]
future = north_america[226:251]
world = gni2[173:173]
world_1973 = gni2[173:174]
world_1994 = gni2[194:195]
world_2014 = gni2[214:215]

```

```python
def per_of_change(country):
    min_before = country.iloc[0]
    max_before = country.iloc[-1]
    per_of_growth = (max_before - min_before)/min_before
    return round(per_of_growth*100, 2)


after_nafta_change = [per_of_change(after_nafta[country]) for country in after_nafta]
before_nafta_change = [per_of_change(before_nafta[country]) for country in before_nafta] 
before_nafta_change
```

```python
plt.plot(after_nafta.index, after_nafta['United States'], color='red', linewidth=2.25, label='USA')
plt.plot(after_nafta.index, after_nafta['Canada'], color='gold', linewidth=2.25, label='Canada')
plt.plot(after_nafta.index, after_nafta['Mexico'], color='green', linewidth=2.25, label='Mexico')
year = years[194:215]
plt.xticks(ticks=after_nafta.index, labels=year, rotation='vertical');
plt.xlabel('Year')
plt.ylabel('GNI')
plt.title("GNI in North America between years 1994 - 2014", size=10, loc='left');
plt.legend();
plt.savefig('after_nafta.png')
```

```python
plt.plot(before_nafta.index, before_nafta['United States'], color='red', linewidth=2.25, label='USA')
plt.plot(before_nafta.index, before_nafta['Canada'], color='gold', linewidth=2.25, label='Canada')
plt.plot(before_nafta.index, before_nafta['Mexico'], color='green', linewidth=2.25, label='Mexico')
year = years[173:194]
plt.xlabel('Year')
plt.ylabel('GNI')
plt.title("GNI in North America between years 1994 - present", size=10, loc='left');
plt.xticks(ticks=before_nafta.index, labels=year, rotation='vertical');
plt.title("GNI years 1973-1993", size=10, loc='left');
plt.legend();
plt.savefig('before_nafta.png')
```
