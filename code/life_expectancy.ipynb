```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

life = pd.read_csv('data/life_expectancy.csv', header=[0], index_col=[0])
life.index.name = 'Year'  
life.fillna(0, inplace=True)
lives = life.transpose()
life_expect = lives.replace({'k': '*1e3', 'M': '*1e6' }, regex=True).map(pd.eval).astype(int)
years = list(life_expect.index)
north_america = life_expect[['United States','Mexico','Canada']]
before_nafta = north_america[173:194]
after_nafta = north_america[194:215]
```

```python
after_nafta
```

```python
plt.plot(before_nafta.index, before_nafta['United States'], color='red', linewidth=2.25, label='USA')
plt.plot(before_nafta.index, before_nafta['Canada'], color='gold', linewidth=2.25, label='Canada')
plt.plot(before_nafta.index, before_nafta['Mexico'], color='green', linewidth=2.25, label='Mexico')
year = years[173:194]
plt.xlabel('Year')
plt.ylabel('GNI')
plt.title("GNI between years 1994 - present", size=10, loc='left');
plt.xticks(ticks=before_nafta.index, labels=year, rotation='vertical');
plt.title("Life expectancy in Norther America years 1973-1993", size=10, loc='left');
plt.legend();
plt.savefig('life_before_nafta.png')
```

```python
plt.plot(after_nafta.index, after_nafta['United States'], color='red', linewidth=2.25, label='USA')
plt.plot(after_nafta.index, after_nafta['Canada'], color='gold', linewidth=2.25, label='Canada')
plt.plot(after_nafta.index, after_nafta['Mexico'], color='green', linewidth=2.25, label='Mexico')
year = years[194:215]
plt.xticks(ticks=after_nafta.index, labels=year, rotation='vertical');
plt.xlabel('Year')
plt.ylabel('GNI')
plt.title("life expectancy in North America between years 1994 - 2014", size=10, loc='left');
plt.legend();
plt.savefig('life_after_nafta.png')
```
